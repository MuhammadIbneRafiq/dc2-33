```python

import pandas as pd
import re

# Load the dataset
df = pd.read_csv('../societal_wellbeing_dataset/final_merged_cleaned_lsoa_london_social_dataset.csv')
print(f"Original shape: {df.shape}")

# Define patterns for each category to drop (without the (?i) flags)
patterns = [
    # Religion
    r'religion|faith|christian|buddhist|hindu|jewish|muslim|sikh|no religion',
    # GCSE and education results
    r'gcse|exam|score|attainment',
    # Road casualties
    r'road|casualt|accident|crash|collision',
    # Public transport
    r'public transport|transport access|accessibility|bus|tube|rail|underground',
    # Health
    r'health|illness|disease|hospital|doctor|nhs|medical|life expect|mortality|death',
    # Qualifications
    r'qualification|educated|degree|diploma',
    # Household language
    r'household language|language|speak|english',
    # Country of birth
    r'country of birth|born|foreign|migrant|immigrant',
    # Level A/AS
    r'level a|a level|as level|a/as',
    # Pupil related
    r'pupil|student',
    # Key stage
    r'key stage|keystage|ks1|ks2|ks 1|ks 2',
    # Absence
    r'absence|absent|attendance',
    # Early years foundation
    r'early years|foundation stage|eyfs',
    # Free school meals
    r'free school meal|fsm',
    # Fuel poverty
    r'fuel poverty|energy|gas|electricity',
    # Child poverty
    r'child poverty|children in poverty'
]

# Combine patterns and use case-insensitive flag for the entire pattern
combined_pattern = '|'.join(patterns)

# Filter columns to drop
columns_to_drop = []
for col in df.columns:
    if re.search(combined_pattern, col, re.IGNORECASE):
        columns_to_drop.append(col)

# Get columns to keep
columns_to_keep = [col for col in df.columns if col not in columns_to_drop]

# Create the filtered dataframe
df_cleaned = df[columns_to_keep]

# Save the cleaned dataset
df_cleaned.to_csv('../societal_wellbeing_dataset/final_merged_cleaned_lsoa_london_social_dataset_reduced.csv', index=False)

print(f"Original columns: {df.shape[1]}")
print(f"Columns kept: {len(columns_to_keep)}")
print(f"Columns removed: {len(columns_to_drop)}")
print(f"Final shape: {df_cleaned.shape}")
```

```python
# Let's check if we have burglary data and use it for the visualization
if 'Crime (numbers);Burglary;2012/13' in lsoa_data.columns:
    value_column = 'Crime (numbers);Burglary;2012/13'
elif 'Crime (rates);Burglary;2012/13' in lsoa_data.columns:
    value_column = 'Crime (rates);Burglary;2012/13'
else:
    # Use a different column if burglary data is not available
    # We'll look for any numeric column that might be useful for visualization
    numeric_cols = lsoa_data.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        if lsoa_data[col].notna().sum() > len(lsoa_data) * 0.8:  # At least 80% non-null values
            value_column = col
            break
    else:
        # If no suitable column is found, use the Index of Multiple Deprivation if available
        if 'The Indices of Deprivation 2010;Domains and sub-domains;IMD Score' in lsoa_data.columns:
            value_column = 'The Indices of Deprivation 2010;Domains and sub-domains;IMD Score'
        else:
            # Otherwise use a simple count
            lsoa_data['count'] = 1
            value_column = 'count'

print(f"Using '{value_column}' for visualization.")

# Try to load and merge all borough shapefiles
print("Loading London borough shapefiles...")
all_boroughs_gdf = None

# Get all shapefile paths
shp_files = glob.glob(os.path.join(shapefile_dir, "*.shp"))

for shp_file in shp_files:
    try:
        borough_name = os.path.basename(shp_file).replace(".shp", "")
        print(f"Processing {borough_name}...")
        
        # Load shapefile
        gdf = gpd.read_file(shp_file)
        
        # Check if the GeoDataFrame contains LSOA information with modern column naming
        has_lsoa = 'lsoa21cd' in gdf.columns
        
        if has_lsoa:
            print(f"Found LSOA information in {borough_name} shapefile")
            if all_boroughs_gdf is None:
                all_boroughs_gdf = gdf
            else:
                all_boroughs_gdf = pd.concat([all_boroughs_gdf, gdf])
        else:
            print(f"Warning: {borough_name} shapefile does not contain LSOA information.")
    except Exception as e:
        print(f"Error processing {shp_file}: {e}")

if all_boroughs_gdf is None or len(all_boroughs_gdf) == 0:
    print("No valid borough shapefiles found with LSOA information.")
    print("Creating a basic map without LSOA boundaries...")
    
    # Create a basic map of London
    m = folium.Map(location=[51.5074, -0.1278], zoom_start=10)
    
    # Add a title
    title_html = '''
    <h3 align="center" style="font-size:16px"><b>London Societal Wellbeing Map</b></h3>
    <h4 align="center" style="font-size:12px"><i>No LSOA boundaries available</i></h4>
    '''
    m.get_root().html.add_child(folium.Element(title_html))
    
    # Save the map
    map_file = 'london_societal_wellbeing_map.html'
    m.save(map_file)
    print(f"Basic map saved as {map_file}")
else:
    print(f"Successfully loaded {len(all_boroughs_gdf)} LSOA boundaries.")
    
    # Now we know the LSOA column is 'lsoa21cd'
    lsoa_col = 'lsoa21cd'
    print(f"Using LSOA column: {lsoa_col}")
    
    # Convert LSOA codes if needed (check if they match format)
    if lsoa_data['Lower Super Output Area'].iloc[0].startswith('E') and not all_boroughs_gdf[lsoa_col].iloc[0].startswith('E'):
        # Format conversion may be needed here if formats don't match
        print("LSOA code format mismatch, attempting to harmonize...")
    
    # Merge the LSOA data with the shapefile data
    merged_gdf = all_boroughs_gdf.merge(
        lsoa_data, 
        left_on=lsoa_col, 
        right_on='Lower Super Output Area',
        how='left'
    )
    
    # Check if the merge worked
    match_count = merged_gdf[value_column].notna().sum()
    print(f"Matched {match_count} out of {len(all_boroughs_gdf)} LSOAs with data.")
    
    if match_count > 0:
        # Create a Folium map
        m = folium.Map(location=[51.5074, -0.1278], zoom_start=10)
        
        # Create a colormap
        colormap = LinearSegmentedColormap.from_list(
            'custom_cmap', ['#ffffcc', '#ffeda0', '#fed976', '#feb24c', '#fd8d3c', '#fc4e2a', '#e31a1c', '#bd0026', '#800026']
        )
        
        # Add a choropleth layer
        folium.Choropleth(
            geo_data=merged_gdf,
            name='choropleth',
            data=merged_gdf,
            columns=[lsoa_col, value_column],
            key_on=f'feature.properties.{lsoa_col}',
            fill_color='YlOrRd',
            fill_opacity=0.7,
            line_opacity=0.2,
            legend_name=value_column
        ).add_to(m)
        
        # Add hover functionality
        style_function = lambda x: {'fillColor': '#ffffff', 
                                    'color': '#000000', 
                                    'fillOpacity': 0.1, 
                                    'weight': 0.1}
        highlight_function = lambda x: {'fillColor': '#000000', 
                                       'color': '#000000', 
                                       'fillOpacity': 0.5, 
                                       'weight': 0.3}
        
        # Ensure LSOA name column is available for tooltips
        lsoa_name_col = 'lsoa21nm' if 'lsoa21nm' in merged_gdf.columns else lsoa_col
        
        folium.features.GeoJson(
            merged_gdf,
            style_function=style_function,
            control=False,
            highlight_function=highlight_function,
            tooltip=folium.features.GeoJsonTooltip(
                fields=[lsoa_col, lsoa_name_col, value_column],
                aliases=['LSOA Code', 'LSOA Name', value_column],
                style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;")
            )
        ).add_to(m)
        
        # Add a title
        title_html = f'''
        <h3 align="center" style="font-size:16px"><b>London {value_column} by LSOA</b></h3>
        '''
        m.get_root().html.add_child(folium.Element(title_html))
        
        # Save the map
        map_file = 'london_lsoa_map.html'
        m.save(map_file)
        print(f"Interactive map saved as {map_file}")
        
        # Create a static plot for demonstration
        fig, ax = plt.subplots(1, 1, figsize=(15, 10))
        merged_gdf.plot(column=value_column, cmap='YlOrRd', linewidth=0.1, ax=ax, edgecolor='0.8', legend=True)
        ax.set_title(f'London {value_column} by LSOA')
        ax.set_axis_off()
        plt.tight_layout()
        
        # Save the static map
        static_map_file = 'london_lsoa_map_static.png'
        plt.savefig(static_map_file, dpi=300, bbox_inches='tight')
        print(f"Static map saved as {static_map_file}")
    else:
        print("Failed to match LSOAs with data. Trying with a different approach...")
        
        # Create a map with borough boundaries instead since we can't match LSOAs
        print("Creating a map with borough boundaries...")
        
        # First, dissolve by borough
        borough_col = 'lad22nm'
        boroughs_gdf = all_boroughs_gdf.dissolve(by=borough_col).reset_index()
        
        # Aggregate data by borough
        borough_data = []
        for borough in boroughs_gdf[borough_col]:
            lsoas_in_borough = all_boroughs_gdf[all_boroughs_gdf[borough_col] == borough][lsoa_col].tolist()
            # Match these LSOAs in our data
            matched_data = lsoa_data[lsoa_data['Lower Super Output Area'].isin(lsoas_in_borough)]
            
            if not matched_data.empty and value_column in matched_data.columns:
                value = matched_data[value_column].mean()
            else:
                value = None
                
            borough_data.append({'Borough': borough, 'Value': value})
        
        # Create borough dataframe
        borough_df = pd.DataFrame(borough_data)
        
        # Merge with the boroughs GeoDataFrame
        merged_boroughs = boroughs_gdf.merge(borough_df, left_on=borough_col, right_on='Borough', how='left')
        
        # Now create a map using borough boundaries
        m = folium.Map(location=[51.5074, -0.1278], zoom_start=10)
        
        # Check if we have any valid data to show
        if merged_boroughs['Value'].notna().sum() > 0:
            # Add a choropleth layer
            folium.Choropleth(
                geo_data=merged_boroughs,
                name='choropleth',
                data=merged_boroughs,
                columns=[borough_col, 'Value'],
                key_on=f'feature.properties.{borough_col}',
                fill_color='YlOrRd',
                fill_opacity=0.7,
                line_opacity=0.5,
                legend_name=f'Average {value_column} by Borough'
            ).add_to(m)
            
            # Add hover functionality
            folium.features.GeoJson(
                merged_boroughs,
                style_function=lambda x: {'fillColor': '#ffffff', 'color': '#000000', 'fillOpacity': 0.1, 'weight': 0.5},
                control=False,
                highlight_function=lambda x: {'fillColor': '#000000', 'color': '#000000', 'fillOpacity': 0.5, 'weight': 1},
                tooltip=folium.features.GeoJsonTooltip(
                    fields=[borough_col, 'Value'],
                    aliases=['Borough', f'Average {value_column}'],
                    style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;")
                )
            ).add_to(m)
            
            title = f'London Borough Average {value_column}'
        else:
            # Just show borough boundaries without data
            for _, row in merged_boroughs.iterrows():
                folium.GeoJson(
                    row['geometry'],
                    style_function=lambda x: {'fillColor': '#cccccc', 'color': '#000000', 'fillOpacity': 0.1, 'weight': 0.7},
                ).add_to(m)
                folium.Marker(
                    location=[row['geometry'].centroid.y, row['geometry'].centroid.x],
                    popup=row[borough_col],
                    icon=folium.DivIcon(html=f"""<div style="font-size: 8pt">{row[borough_col]}</div>""")
                ).add_to(m)
            
            title = 'London Borough Boundaries'
        
        # Add a title
        title_html = f'''
        <h3 align="center" style="font-size:16px"><b>{title}</b></h3>
        '''
        m.get_root().html.add_child(folium.Element(title_html))
        
        # Save the map
        map_file = 'london_borough_map.html'
        m.save(map_file)
        print(f"Borough map saved as {map_file}")

print("donenenfydfyd") 
```

```python
def load_police_force_tables(excel_file):
    """
    Load police force area tables from Excel file
    """
    print(f"Loading data from {excel_file}...")
    
    # Use openpyxl engine to load the Excel file
    xlsx = pd.ExcelFile(excel_file, engine='openpyxl')
    
    # Print sheet names to identify which contain burglary data
    sheet_names = xlsx.sheet_names
    print(f"Available sheets: {sheet_names}")
    
    # Try to find sheets that might contain burglary data
    burglary_sheets = [name for name in sheet_names if re.search(r'burglary|crime|offence', name.lower())]
    
    if not burglary_sheets:
        print(f"No sheets found with 'burglary' in the name. Using all available sheets.")
        burglary_sheets = sheet_names
    
    # Load data from each relevant sheet
    dfs = {}
    for sheet in burglary_sheets:
        try:
            df = pd.read_excel(xlsx, sheet_name=sheet)
            print(f"Loaded sheet '{sheet}' with {len(df)} rows and {len(df.columns)} columns")
            # Store with sheet name as key
            dfs[sheet] = df
        except Exception as e:
            print(f"Error loading sheet '{sheet}': {e}")
    
    return dfs

def load_offence_classification(excel_file):
    """
    Load offence group classification from Excel file
    """
    print(f"Loading offence classification from {excel_file}...")
    
    try:
        xlsx = pd.ExcelFile(excel_file, engine='openpyxl')
        sheet_names = xlsx.sheet_names
        print(f"Available sheets: {sheet_names}")
        
        # Try to find the main sheet with classification data
        df = pd.read_excel(xlsx, sheet_name=0)  # Default to first sheet
        print(f"Loaded offence classification with {len(df)} rows and {len(df.columns)} columns")
        return df
    except Exception as e:
        print(f"Error loading offence classification: {e}")
        return None

def load_street_level_data(directory_path):
    """
    Load street level crime data from CSV files
    """
    print(f"Loading street level data from {directory_path}...")
    
    # Find all CSV files in the directory
    csv_files = glob.glob(os.path.join(directory_path, "*.csv"))
    
    # Filter for street level files
    street_files = [f for f in csv_files if "street" in f.lower()]
    outcome_files = [f for f in csv_files if "outcome" in f.lower()]
    
    print(f"Found {len(street_files)} street level files and {len(outcome_files)} outcome files")
    
    # Load and combine street data
    street_data = []
    for file in street_files:
        try:
            df = pd.read_csv(file)
            print(f"Loaded {os.path.basename(file)} with {len(df)} rows")
            street_data.append(df)
        except Exception as e:
            print(f"Error loading {file}: {e}")
    
    # Combine all street data files
    if street_data:
        combined_street_data = pd.concat(street_data, ignore_index=True)
        print(f"Combined street data has {len(combined_street_data)} rows")
    else:
        combined_street_data = pd.DataFrame()
        print("No street data loaded")
    
    # Load and combine outcome data
    outcome_data = []
    for file in outcome_files:
        try:
            df = pd.read_csv(file)
            print(f"Loaded {os.path.basename(file)} with {len(df)} rows")
            outcome_data.append(df)
        except Exception as e:
            print(f"Error loading {file}: {e}")
    
    # Combine all outcome data files
    if outcome_data:
        combined_outcome_data = pd.concat(outcome_data, ignore_index=True)
        print(f"Combined outcome data has {len(combined_outcome_data)} rows")
    else:
        combined_outcome_data = pd.DataFrame()
        print("No outcome data loaded")
    
    return combined_street_data, combined_outcome_data

def extract_residential_burglary(street_data, offence_classification=None):
    """
    Extract residential burglary records from street level data
    """
    print("Extracting residential burglary records...")
    
    # Check if data is empty
    if street_data.empty:
        print("No street data available for extraction")
        return pd.DataFrame()
    
    # Print available columns
    print(f"Available columns: {street_data.columns.tolist()}")
    
    # Find the column that contains crime type information
    crime_type_col = None
    for col in ['Crime type', 'crime_type', 'Offence_type', 'offence_type']:
        if col in street_data.columns:
            crime_type_col = col
            break
    
    if not crime_type_col:
        print("Could not find crime type column in the data")
        return pd.DataFrame()
    
    print(f"Using column '{crime_type_col}' for crime type information")
    
    # Extract unique crime types for reference
    unique_crime_types = street_data[crime_type_col].unique()
    print(f"Unique crime types in data: {unique_crime_types}")
    
    # Define terms to identify residential burglary
    residential_terms = ['residential', 'dwelling', 'home', 'house', 'apartment', 'flat']
    
    # First try to find records explicitly marked as residential burglary
    mask = street_data[crime_type_col].str.lower().str.contains('burglary', na=False)
    burglary_data = street_data[mask].copy()
    
    # If we have separate residential/non-residential categorization
    if any(street_data[crime_type_col].str.lower().str.contains('|'.join(residential_terms), na=False)):
        # Further filter for residential burglaries
        residential_mask = burglary_data[crime_type_col].str.lower().str.contains('|'.join(residential_terms), na=False)
        residential_burglary = burglary_data[residential_mask].copy()
        print(f"Found {len(residential_burglary)} explicitly residential burglary records")
    else:
        # If no explicit residential marking, just use all burglary records
        residential_burglary = burglary_data
        print(f"No explicit residential marking found, using all {len(residential_burglary)} burglary records")
    
    # Check if we have LSOA information
    lsoa_col = None
    for col in ['LSOA code', 'LSOA_code', 'lsoa_code', 'LSOA']:
        if col in residential_burglary.columns:
            lsoa_col = col
            break
    
    if lsoa_col:
        print(f"LSOA information available in column '{lsoa_col}'")
        # Count burglaries by LSOA
        lsoa_counts = residential_burglary.groupby(lsoa_col).size().reset_index(name='count')
        print(f"Found {len(lsoa_counts)} LSOAs with burglary data")
    else:
        print("No LSOA information available in the data")
    
    return residential_burglary


# Set paths to data files
pfa_tables_2023 = "pfatablesyedec23.xlsx"
pfa_tables_2024 = "policeforceareatablesyedecember2024.xlsx"
offence_classification = "offence-group-classification-june-2022.xlsx"
street_data_dir = os.path.join("6b551cfc75574fc866735197ccffb4296954d23a", "2025-03")

# Create output directory if it doesn't exist
output_dir = "burglary_analysis"
os.makedirs(output_dir, exist_ok=True)

# Load data from Excel files
pfa_dfs_2023 = load_police_force_tables(pfa_tables_2023)
pfa_dfs_2024 = load_police_force_tables(pfa_tables_2024)
offence_df = load_offence_classification(offence_classification)

# Load street level data
street_data, outcome_data = load_street_level_data(street_data_dir)

# Extract residential burglary records
residential_burglary = extract_residential_burglary(street_data, offence_df)

# If residential burglary data is available
if not residential_burglary.empty:
    # Save the extracted data
    output_file = os.path.join(output_dir, "residential_burglary_by_lsoa.csv")
    residential_burglary.to_csv(output_file, index=False)
    print(f"Saved residential burglary data to {output_file}")
    
    # Analyze burglary by LSOA
    if 'LSOA code' in residential_burglary.columns:
        lsoa_summary = residential_burglary.groupby(['LSOA code', 'LSOA name']).size().reset_index(name='burglary_count')
        lsoa_summary = lsoa_summary.sort_values('burglary_count', ascending=False)
        
        # Save LSOA summary
        lsoa_summary_file = os.path.join(output_dir, "lsoa_burglary_summary.csv")
        lsoa_summary.to_csv(lsoa_summary_file, index=False)
        print(f"Saved LSOA burglary summary to {lsoa_summary_file}")
        
        # Print top 10 LSOAs with highest burglary counts
        print("\nTop 10 LSOAs with highest residential burglary counts:")
        print(lsoa_summary.head(10))

print("\nData processing complete!")
```
